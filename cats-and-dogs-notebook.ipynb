```python
import tensorflow as tf 
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Preprocessing the dataset
training_data_generator = ImageDataGenerator(
                          rescale=1./255,
                          shear_range=0.2,
                          zoom_range=0.2,
                          horizontal_flip = True
)
train_generator = training_data_generator.flow_from_directory(
                    '/kaggle/input/dogs-cats-images/dataset/training_set',
                    target_size=(64, 64),
                    batch_size=32,
                    class_mode = 'binary'
)
```

```python
#Preprocessing test set
test_set_generator = ImageDataGenerator(
                        rescale=1/255
)
test_data = test_set_generator.flow_from_directory(
            '/kaggle/input/dogs-cats-images/dataset/test_set',
            target_size=(64,64),
            batch_size=32,
            class_mode='binary'
)
```

```python
# Building the CNN

cnn = tf.keras.models.Sequential()
# convolution
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
# pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
# Adding the second concolutional layer
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
# Flattening the data
cnn.add(tf.keras.layers.Flatten())
# Adding Artificial Neural network
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
# output layer
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# compiling the cnn
cnn.compile(optimizer='adam', metrics='accuracy', loss='binary_crossentropy')
```

```python
# fitting the data to the neural net
cnn.fit(x=train_generator, validation_data=test_data, epochs=35)
```

```python
# Making a prediction
import numpy as np
import keras.utils as image
test_image = image.load_img('/kaggle/input/single-prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
```

```python
result = cnn.predict(test_image)
train_generator.class_indices
```
